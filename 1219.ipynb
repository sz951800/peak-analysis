# WLS

```python
import numpy as np
import pandas as pd


# read all dataframes
peak1=pd.read_csv('/Users/shuchanzhou/Desktop/CSV_peaks/No Filter Full gain - singlets/Peak 1.csv').drop(columns='index', errors='ignore')
peak2=pd.read_csv('/Users/shuchanzhou/Desktop/CSV_peaks/No Filter Full gain - singlets/Peak 2.csv').drop(columns='index', errors='ignore')
peak3=pd.read_csv('/Users/shuchanzhou/Desktop/CSV_peaks/No Filter Full gain - singlets/Peak 3.csv').drop(columns='index', errors='ignore')
peak4=pd.read_csv('/Users/shuchanzhou/Desktop/CSV_peaks/No Filter Full gain - singlets/Peak 4.csv').drop(columns='index', errors='ignore')
peak5=pd.read_csv('/Users/shuchanzhou/Desktop/CSV_peaks/No Filter Full gain - singlets/Peak 5.csv').drop(columns='index', errors='ignore')
peak6=pd.read_csv('/Users/shuchanzhou/Desktop/CSV_peaks/No Filter Full gain - singlets/Peak 6.csv').drop(columns='index', errors='ignore')
peak7=pd.read_csv('/Users/shuchanzhou/Desktop/CSV_peaks/No Filter Full gain - singlets/Peak 7.csv').drop(columns='index', errors='ignore')
peak8=pd.read_csv('/Users/shuchanzhou/Desktop/CSV_peaks/No Filter Full gain - singlets/Peak 8.csv').drop(columns='index', errors='ignore')


# a list of dataframes (each corresponding to a peak)
peaks = [peak1, peak2, peak3, peak4, peak5, peak6, peak7, peak8] 

# a list of all 64 detectors
detectors=list(peak1.columns)

results = []

for detector in detectors:
    e = []
    v = []
    N = []
    weights = []
    
    for peak in peaks:
        e_i = peak[detector].mean()
        v_i = peak[detector].var()
        N_i = len(peak[detector])
        w_i = (N_i - 1) / (2 * e_i**2)
        e.append(e_i)
        v.append(v_i)
        N.append(N_i)
        weights.append(w_i)
    
    # Convert lists to numpy arrays
    e = np.array(e)
    v = np.array(v)
    N = np.array(N)
    weights = np.array(weights)

    # Design matrix, each column 
    X = np.column_stack((np.ones_like(e), e, e**2))

    # Diagonal weight matrix
    W = np.diag(weights)

    # Weighted least squares solution
    XTW = X.T @ W
    XTWX = XTW @ X
    XTWY = XTW @ v
    beta = np.linalg.solve(XTWX, XTWY)
    #beta = np.linalg.inv(XTWX) @ XTWY
    # Coefficients
    baseline, gain, cv_2 = beta
    cv=np.sqrt(cv_2)
    results.append([baseline, gain, cv])

# Create a DataFrame with results
df_results = pd.DataFrame(results, columns=['baseline', 'gain', 'CV'], index=detectors)
```
